```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,24])
y = np.array([2,4])
```

```python
plt.plot(x,y)
font ={'family':'serif','color':'blue','size':20}
plt.title('Student_Mark',fontdict=font) #(loc=) to locate the text
plt.xlabel('Age')
plt.ylabel('mark')
plt.show()
```

##Without line plot

```python
plt.plot(x,y,'o')
plt.show()
```

##Multiple points and subplot

```python
X = np.array([5,10,15,20,25])
Y = np.array([2,7,4,9,5])
```

```python
a=np.array([2,3,4,5,6,8,9,7])
b=np.array([3,4,6,8,10,5,1,11])
```

```python
plt.subplot(1,2,1)
plt.plot(X,Y)
plt.subplot(1,2,2)
plt.plot(a,b)
plt.suptitle('Ploting two graph')
plt.show()
```

##Scatter plot  

```python
sizes=np.array([2500,70,55,60,85,95,125,135])
plt.scatter(a,b,s=sizes,alpha=0.5)
plt.show()
```

##Combine scatter plot

```python
plt.scatter(x,y)#color
plt.scatter(a,b)
plt.show()
```

```python
colors=np.array(['red','blue','green','black','yellow','hotpink','orange','purple'])
```

```python
plt.scatter(a,b,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

##Bars

```python
a1=np.array(['A','B','C','D'])
b2=np.array([3,6,7,9])
```

```python
plt.bar(a1,b2,color='red')
plt.show()
```

```python
plt.barh(a1,b2)
plt.show()
```

```python
b1=np.array([5,10,15,20])
plt.hist(b1)
plt.show()
```

```python
a2=np.array([25,30,15,20,10])
my_labels =["Tiger","Lion","Cheetha","Wolf","Elephant"]
plt.pie(a2,labels=my_labels,startangle=90)
plt.legend(title="Animals")
plt.show()
```

```python
myexplode =[0.2,0,0,0,0]
plt.pie(a2, labels=my_labels, explode=myexplode, shadow=True)#shadow=true
plt.show()
```
